# Assignment 1 Q5 - Henon Map

## Part d

```python
# set parameters
alpha = 1.25;
beta = 0.3;

p = np.array([alpha,beta])

# set number of iterations
Nsteps = 1000;

# set initial condition
x0 = np.array([0.5,0.2]);
```

```python
# create empty solution vector 
xtraj = np.empty((x0.shape[0],Nsteps+1));
xtraj[:] = np.nan

# input initial condition
xtraj[:,0] = x0;
```

### Iterate map

```python
for i in np.arange(0,Nsteps):
    x0 = HenonMap(x0,p);
    xtraj[:,i+1] = np.squeeze(x0);
```

### Plot solution

```python
fig, axs = plt.subplots(1, 2, layout='constrained', figsize=(10, 5))

ax1 = axs[0]
ax1.plot(xtraj[0,200:],xtraj[1,200:],'.')
ax1.set_xlabel("$x_n$")
ax1.set_ylabel("$y_n$")
ax1.set_title("Solution Trajectory in Phase Plane")
ax1.set_xlim([-2, 2])
ax1.set_ylim([-0.5, 0.5])

ax2 = axs[1] 
ax2.plot(np.arange(0,Nsteps+1),xtraj[0,:],'.')
ax2.set_xlabel("$n$")
ax2.set_ylabel("$x_n$")
ax2.set_title("Solution Trajectory ($x_n$)")
ax2.set_xlim([0, 1000])

plt.show()
```

### Interpretation
We can see from the phase plane plot above that the solution converges to a period 8 solution.  This is further suported by right figure, where we see convergence shortly after 100 iterations.
