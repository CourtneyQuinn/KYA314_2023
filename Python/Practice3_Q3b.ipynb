# Practice 3 Q3b - Vander Pol Oscillator

## $\alpha < 0$

```python
# set parameters
alpha = -1.

def VO_model(t,x):
    xout = VanderPol(t,x,alpha)
    return xout
```

```python
# set initial conditions
x0 = np.array([0.1,0.1]);
```

```python
# IVP solver settings
tspan = [0,100]
h = 0.01

xout,t,xend = MyIVP(VO_model,x0,tspan,h)

xout = np.squeeze(xout)
```

### Plot solution in phase plane

```python
plt.plot(xout[0,:],xout[1,:])
plt.xlabel("$x$")
plt.ylabel("$\dot{x}$")
plt.title("Solution Trajectory in Phase Plane")
plt.xlim([-0.2, 0.2])
plt.ylim([-0.2, 0.2])

plt.show()
```
